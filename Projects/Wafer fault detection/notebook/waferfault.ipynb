### 9. Handling the Nan values

#### Extracting columns with nan values

```python
nan_columns = df.columns[df.isna().any()].tolist()
```

```python
nan_columns
```

```python
for column in nan_columns:
    median_val = df[column].median()
    df[column].fillna(median_val,inplace=True)
```

### 12. Model Training

models

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

evaluation matrics

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

Evaluation matrics

```python
def evaluate_model(actual,predicted):
    mse = mean_squared_error(actual,predicted)
    mae = mean_absolute_error(actual,predicted)
    rmse = np.sqrt(mean_squared_error(actual,predicted))
    r2 = r2_score(actual,predicted)
    r2_adjusted = 1 - ((1-r2_score(actual,predicted))*(len(y)-1))/(len(y)- X.shape[1] - 1)

    return mse,mae,rmse,r2,r2_adjusted
```

```python
models = {
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'ElasticNet':ElasticNet()
    
}

R2_list = []
R2_adjusted_list = []
model_list = []

for i in range(len(list(models))):
    model = list(models.values())[i]

    model.fit(X_train,y_train)

    #Make Predictions
    y_pred = model.predict(X_test)
    mse,mae,rmse,r2,r2_adjusted = evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print("Model Traing Performance")
    print('MSE',mse)
    print('MAE',mae)
    print('RMSE',rmse)
    print('R2',r2*100)
    print('R2_adjusted',r2_adjusted*100)

    R2_list.append(r2)
    R2_adjusted_list.append(r2_adjusted)
    
    print('='*40)
    print('\n\n')
```
