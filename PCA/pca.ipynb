```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
import pandas 
%matplotlib inline
```

```python
#import data
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df = pandas.DataFrame(cancer_dataset['data'], columns= cancer_dataset['feature_names'])
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#Standardize the data(center the data)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
centered_data = scaler.transform(df)
```

```python
centered_data
```

```python
## Applying PCA algorithm
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components= 2)
```

```python
pca_data = pca.fit_transform(centered_data)
```

```python
pca_data
```

```python
#Visulization
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1], c = cancer_dataset['target'],cmap='plasma')
plt.xlabel("First Principle Component")
plt.ylabel("Second Principle Components")
plt.show()
```
